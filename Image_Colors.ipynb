```python
from PIL import Image
from matplotlib.pylab import plt
import numpy as np
```

```python
def ReadImage(path):
    img = Image.open(path)
    c_img = img.convert("P")
    palette = c_img.getpalette()
    a_img = np.array(c_img)
    return a_img,palette
```
